# 3. Bar graph_ male-female buyers

```python
male=cars_data['Buyer Gender'][cars_data['Buyer Gender']=='Male']
female=cars_data['Buyer Gender'][cars_data['Buyer Gender']=='Female']
gender=['Male','Female']
gender_count=[male.count(),female.count()]
plt.figure(figsize=(5,6))
plt.title('male - female buyers')
plt.xlabel('gender')
plt.ylabel('no. of buyers')
plt.bar(gender,gender_count)
```

# 4. Top 5 cars based on sales price

```python
cars_sales=cars_data.sort_values(by='Sale Price',ascending=False)
```

```python
top5Cars=cars_sales.head(5)
```

```python
top5Cars
```

```python
plt.title('top 5 cars based on sales price')
plt.xlabel('model')
plt.ylabel('sales')
plt.bar(top5Cars['Model'],top5Cars['Sale Price'])
```

```python
# top 5 cars based on their sales price are: 
```

```python
top5=[]
for x in top5Cars.index:
            name=cars_data['Make'][x]+' '+cars_data['Model'][x]
            top5.append(name)
print(top5)
```
